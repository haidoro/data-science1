# Lesson3 グラフを描く

## ライブラリのimport

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

## trainデータの読み込み
- read_csv関数でcsvの読み込み

```python
train = pd.read_csv("train.csv")
```

## trainの先頭行を表示

```python
train.head()
```

## trainのyの折れ線グラフを描画
- 折れ線グラフを描くにはplot関数を使います

```python
train["y"].plot()
```

## グラフのサイズを変更
- カッコの中にオプションとしてfigsize=(12,4)と書きます。
12が横4が縦の比率です。

```python
train["y"].plot(figsize=(12,4))
```

## グラフにタイトルを付ける
- カッコの中にオプションとして、title="XXX"と書きます
- カッコの中が複数となった場合はカンマ（,）を使うことで複数書くことができます

```python
train["y"].plot(figsize=(12,4),title="Sales Trends")
```

## グラフのx軸とy軸に名前を付ける
- ちょっと複雑で一旦、plot関数の結果を別の変数に代入する必要があります
- 今回は変数axに代入することにします
- x軸はaxを使い、set_xlabel("name")と書くことで名前をつけることができます

```python
ax = train["y"].plot(figsize=(12,4),title="Sales Trends")
ax.set_xlabel("time")
ax.set_ylabel("num")
```

## trainのyのヒストグラムを描く
- plot.hist関数を使います

```python
train["y"].plot.hist()
```

## グリッド線を引く
- カッコの中にオプションとして、grid=Trueと書きます

```python
train["y"].plot.hist(grid=True)
```

## ヒストグラム上に平均値を表す赤線を描く
- ヒストグラムに平均値を表す線を描画
- まず縦線の描き方から始めます。縦線を描くときにはplt.axvline(x=NUMBER, color="red")と書きます
- vlineのvはvertical（垂直）

```python
plt.axvline(x=train["y"].mean(),color="red")
train["y"].plot.hist()
```

## 上記のヒストグラムを大きさ(12,4)にし、タイトルを"ヒストグラム"として、pngファイルの図として保存
- plt.savefig("FILE_NAME.拡張子")を一番最後に記述
- ファイル名は「sample.png」

```python
plt.axvline(x=train["y"].mean(),color="red")
train["y"].plot.hist(figsize=(12,4),title="ヒストグラム")
plt.savefig("sample1.png")
```

## yとweekの関係を調べる為に、箱ひげ図を描く
- 箱ひげ図の描き方は、yとweekを選択した上で、boxplot(by="week")と書きます
- by="week"はx軸にしたいカラムを選択します
- ○は外れ値です

```python
train[["y","week"]].boxplot(by="week")
```

# 練習

trainのtemperatureの折れ線グラフを描画。タイトルとx軸、y軸にもそれぞれ名前をつける。
- x軸とy軸に名前を付ける為に、グラフは一旦変数axに代入
- タイトル名は「temperature」、x軸は「time」、y軸は「temperature」

- x軸はweather

```python
train[["y","weather"]].boxplot(by="weather")
```
